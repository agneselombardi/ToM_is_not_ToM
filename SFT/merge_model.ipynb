```python
# Merge LoRA adapter into base model
from transformers import AutoTokenizer, AutoModelForCausalLM
from peft import PeftModel

base_model_id = "meta-llama/Meta-Llama-3-8B"
base_model_cache = "/extra/agnese.lombardi/ISA_ToM"
base_model = AutoModelForCausalLM.from_pretrained(base_model_id, cache_dir=base_model_cache, torch_dtype="auto")

tokenizer = AutoTokenizer.from_pretrained(base_model_id, cache_dir=base_model_cache)

lora_adapter_path = "/extra/agnese.lombardi/ISA_ToM/ToM_fine_tuned"
model = PeftModel.from_pretrained(base_model, lora_adapter_path)

model = model.merge_and_unload(safe_merge=True)

# Save as a new full model
merged_path = "/extra/agnese.lombardi/ISA_ToM/tom_model"
model.save_pretrained(merged_path)
tokenizer.save_pretrained(merged_path)
```
